```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import os
```

```python
import pandas as pd

# Load your dataset
df = pd.read_csv('/content/Road Accident Data.csv')  # Replace with your actual dataset file

# Inspect the data
print(df.info())  # Check for missing values and data types
print(df.head())  # Display first few rows of the dataset
# Check for missing values
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print(missing_values)

# Drop rows with missing values (you can also impute them based on the context)
df = df.dropna(subset=[ 'Weather_Conditions', 'Road_Surface_Conditions', 'Time'])
```

```python
# Convert 'Time' column to datetime and extract the hour
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M', errors='coerce')
df['Hour'] = df['Time'].dt.hour
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot accidents by road surface conditions
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Road_Surface_Conditions')
plt.title('Accidents by Road Surface Conditions')
plt.xlabel('Road Surface Condition')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
# Plot accidents by weather conditions
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Weather_Conditions')
plt.title('Accidents by Weather Conditions')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
# Plot accidents by hour of day
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Hour')
plt.title('Accidents by Time of Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Accidents')
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the plotting environment
fig, axs = plt.subplots(1, 2, figsize=(16, 6))

# Plot accidents by weather conditions
sns.countplot(y='Weather_Conditions', data=df, ax=axs[0],
              order=df['Weather_Conditions'].value_counts().index,
              palette='Blues_r')
axs[0].set_title('Accidents by Weather Conditions')
axs[0].set_xlabel('Number of Accidents')
axs[0].set_ylabel('Weather Condition')

# Plot accidents by road surface conditions
sns.countplot(y='Road_Surface_Conditions', data=df, ax=axs[1],
              order=df['Road_Surface_Conditions'].value_counts().index,
              palette='Greens_r')
axs[1].set_title('Accidents by Road Surface Conditions')
axs[1].set_xlabel('Number of Accidents')
axs[1].set_ylabel('Road Surface Condition')

plt.tight_layout()
plt.show()
```
